# REPLACE DEĞER DEĞİŞTİRME

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("nba.csv")
```

```python
df
```

```python
def yas(oyuncu_yas):
    if oyuncu_yas < 25 :
        return "1"
    elif (oyuncu_yas >= 25) & (oyuncu_yas < 30) :
        return "2"
    else :
        return "3"
```

```python
df["oyuncu_yas"] = df["Age"].apply(yas)
```

```python
df
```

```python
df["oyuncu_yas"].replace(to_replace = "1" , value = "Genç Oyuncu")
```

```python
df.head(40)
```

```python
df["oyuncu_yas"] = df["oyuncu_yas"].replace(to_replace = "1" , value = "Genç Oyuncu")
```

```python
df.head(25)
```

```python
df.drop("oyuncu_yas" , axis = 1 , inplace = True)
```

# MAP İLE DEĞER DEĞİŞTİRME

```python
def oyuncu_maas(maas):
    if maas < 1000000 : 
        return "1"
    elif (maas >= 1000000) & (maas < 8000000) :
        return "2"
    else :
        return "3"
```

```python
df["oyuncu_maas"] = df["Salary"].apply(oyuncu_maas)
```

```python
df
```

```python
map_1 = {
    "1" : "Düşük Maaş",
    "2" : "Orta Düzey Maaş",
    "3" : "Yüksek Maaş"
}
```

```python
df["oyuncu_maas"].map(map_1)
```

```python
df["oyuncu_maas"] = df["oyuncu_maas"].map(map_1)
```

# groupby kullanımı

```python
df.head(20)
```

```python
# ilk olarak aracı markalarına göre gruplandırıp ortalamalara bakalım

df.groupby("make").mean().head(20)
```

```python
# şimdide toplamları görelim 

df.groupby("make").sum().head(20)
```

```python
# şimdi vitese göre gruplandırma yapalım ve açıklayıcı istatistikleri görelim ...

df.groupby("year").describe().transpose() # transpose komutu ile transpozunu aldık tablonun
```

```python
# araçları hem yakıt hemde vitese göre gruplandırıyoruz ve ayrı bir dataframe olarak alıyoruz ... 

arac_bilgi = df.groupby(["fuel","gear"]).describe()
```

# AGGREGATE

```python
# aşağıdaki kod ile dosyamızdaki min max ve ortalama değerleri hızlıca görebiliriz ...

df.agg(["min","max","mean"])
```

```python
# hangi sütunları istiyorsak onları çağırabiliriz ...

df.agg(["min","max","mean"])[["mileage","hp"]]
```

```python
# hangi sütünla ilgili istatistiksel değerleri görmek istediğimizi ayarlayabiliriz ...
 
df.agg({
    "price": ["mean","min"] , 
    "year" : ["max" , "mean"]
})
```

```python
# yukardaki işlemi gruplandırma yaparakta kullanabiliriz ...

df.groupby("gear").agg({
    "price": ["mean","min"] , 
    "year" : ["max" , "mean"]
}
)
```

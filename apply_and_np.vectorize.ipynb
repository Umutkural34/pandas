Var olan değişkenin içerisinde değişiklik yapabilir ya da DataFrame objesi içerisinde yer alan değişkenlerden yeni değişkenler türetilebilir.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("nba.csv")
```

```python
df
```

```python
# oyuncuların ilk 5 harfinin olduğu bir sütun oluşturalım ve bunun için önce bir fonksiyon oluşturalım ve apply komutuna bu
# fonksiyonu atayalım ...

def ilk_bes_harf(oyuncu_ad):
    return str(oyuncu_ad)[:5]
```

```python
# oyuncu_ilk_5_harf adında yeni bir sütun oluşturuyorum ve name sütununu kullanarak apply ile bu sütunu ekliyorum ...

df["oyuncu_ilk_5_harf"] = df["Name"].apply(ilk_bes_harf)
```

```python
df.head(20)
```

```python
# bu sütunu siliyoruz ...

df.drop("oyuncu_ilk_5_harf" , axis = 1 , inplace = True)
```

```python
# şimdide oyuncuların yaşlarına göre tecrübelerine yönelik bir sütun eklemek istiyorum ve yine öncelikle 
# bir fonksiyon oluşturuyorum ...

def tecrube(yas):
    if yas < 25 :
        return "Genç Oyuncu"
    elif (yas >= 25) & (yas <= 30) :
        return "Gelişmiş Oyuncu"
    else :
        return "Tecrübeli Oyuncu"
        
```

```python
# oyuncu_tecrube adında yeni bir sütun oluşturuyorum ve age sütununu kullanarak apply ile bu sütunu ekliyorum ...

df["oyuncu_tecrübe"] = df["Age"].apply(tecrube)
```

```python
df.head(50)
```

```python
# az önce oluşturduğum sütunu siliyorum ...

df.drop("oyuncu_tecrübe" , axis = 1 , inplace = True)
```

```python
# yeni bir csv dosyası oluşturuyorum ... (youtube da en çok izlenen videolar)

df = pd.read_csv("GBvideos.csv")
```

```python
df
```

```python
# videonun görüntülenme ve aldığı beğeni sayısına göre bir oranlama yapmak istiyorum

def oran(begeni , goruntulenme):
    return (begeni / goruntulenme).round(3)  
    
```

```python
# oran adında yeni bir sütun oluşturuyorum ve , views ve likes sütunlarını kullanarak apply ile bu sütunu ekliyorum ...

df["Oran"] = df[["likes","views"]].apply(lambda df : oran(df["likes"] , df["views"]) , axis = 1)
```

```python
df.head(25)
```

# np.vectorize

```python
# az önce apply ile oran adlı bir sütun oluşturduk bunu şimdi np.vectorize ile yapalım

def oran2(begeni , goruntulenme):
    return (begeni / goruntulenme) 
    
```

```python
# oran2 adında yeni bir sütun oluşturuyorum ve , views ve likes sütunlarını kullanarak np.vectorize ile bu sütunu ekliyorum ...

df["Oran2"] = np.vectorize(oran2)(df["likes"] , df["views"])
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Covid Live.csv")
```

```python
df
```

```python
# isna veya isnull komutu ile hücrelerin boş veya dolu olduğunu anlayabiliriz ...

# df.isna() (bunu da kullanabiliriz ...)
df.isnull()
```

```python
# notnull ile hangi hücrenin dolu olduğunu yine anlayabiliriz ...

df.notnull()
```

```python
# info komutu ile sütunlarda kaç tane girilmiş veri olduğunu görebiliriz ...

df.info()
```

```python
# isnull.sum ile hangi sütunda kaç girilmemiş veri , isnull.sum.sum komutu ile toplamda kaç eksik veri olduğunu görebiliriz ...

print("Sütunlardaki eksik veriler :" , df.isnull().sum())
print("------------------------------")
print("Toplam eksik veri sayısı : " , df.isnull().sum().sum() )
```

```python
# eksik veri sayısını yüzdesel olarak hesaplayalım ...

toplam_veri = df.size
eksik_veri = df.isnull().sum().sum()

print("Eksik veri : %" ,round(toplam_veri / eksik_veri))
```

```python
# New\nDeaths sütununun dolu olduğu satırları görmek istersek

df[df["New\nDeaths"].notnull()]
```

```python
# belirlediğimiz iki sütununda dolu olduğu verileri görmek istersek ...

df[(df["Serious,\nCritical"].notnull()) & (df["Total\nTests"].notnull())]
```

# DROPNA

```python
# herhangi bir satırda boş değer varsa siliyoruz ...

df.dropna()
```

```python
# herhangi bir sütunda boş değer varsa siliyoruz ...

df.dropna(axis=1)
```

```python
df.info()
```

```python
# thresh komutu ile filtreleme yapabiliriz ...

df.dropna(thresh = 100 , axis = 1) # sütunlarda en az 100 hücre dolu olanları getirir 
```

```python
# subset komutu ile özel olarak ilgilenmek istedğimiz sütunları seçebiliriz ...

df.dropna(subset = ["Total\nTests"]) # sadece Total\nTests sütununda boş olan değerleri kaldırır ...
```

```python
# Population ve Serious,\nCritical sütunları üzerindeki boş değerleri kaldıralım ...

df.dropna(subset = ["Population","Serious,\nCritical"])
```

# TRANSFORM

```python
df
```

```python
# cinsiyetlerin ortalamasına göre mass kısmının ortalamasını alalım ...

df["mass"] = df["mass"].fillna(df.groupby("sex")["mass"].transform("mean"))
```

```python
# türlere ve cinsiyetlere göre gruplandırma yapıp boy kısmını ortalamaya göre yapalım ...

df["height"] = df["height"].fillna(df.groupby(["sex","species"])["height"].transform("mean"))
```

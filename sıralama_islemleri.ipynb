# SIRALAMA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("nba.csv")
```

```python
df
```

```python
# aşağıdaki kod ile oyuncu isimlerini sıralayabiliriz ...

df.sort_values("Name")
```

```python
# ters bir şekilde sıralamak istersek ...

df.sort_values("Name" , ascending = False)
```

```python
# iki sütunda birden sıralama işlemi yapabiliriz ...

df.sort_values(["Team","Weight"]).head(25)
```

# DEĞER SAYMA

```python
# hangi takım kaç kere var görelim 

df.value_counts("Team")
```

```python
# hangi pozisyondan kaç oyuncu olduğunu görelim 

df.value_counts("Position")
```

# EN BÜYÜK VE EN KÜÇÜK DEĞERLER

```python
# en yüksek 8 maaş

df.nlargest(8,"Salary")
```

```python
# en düşük 12 maaş

df.nsmallest(12,"Salary")
```

# SAMPLE

```python
# rastgele 10 satır ...

df.sample(10)
```

### KOŞUL İFADELERİ 

```python
# koşul ifadeleri genel olarak dataframemizin içine aşağıdaki gibi yazılır ... 

df[df["Age"] > 30] # yaşı 30 dan büyük oyuncular
```

```python
# en düşük ve en yüksek maaşı alan oyuncular

print(df[df["Salary"] == df["Salary"].min()]) # en düşük maaş alan oyuncu
print("---------------------------------------")
print(df[df["Salary"] == df["Salary"].max()]) # en yüksek maaş alan oyuncu
```

```python
df[(df["Team"] == "New York Knicks") & (df["Age"] < 30)] # takımı new york olup 30 yaşından küçük olan oyuncular
```

```python
df[(df["Team"] == "Boston Celtics") | (df["Team"] == "Brooklyn Nets")] # takımı boston veya brooklyn olanlar
```

```python
df[df["Team"] != "Utah Jazz"].head(20) #takımı utah olmayan oyuncular
```

```python
a = ["Tyler"]
df["Name"].isin(a) # name kısmında tyler içeren oyuncular
```

# CONCAT NEDİR ?

```python
import pandas as pd
import numpy as np
```

```python
calisan_1 = {
    "calisan_id":[1,2,3,4],
    "calisan_isim":["Aynur Mavi","Ahmet Yıldırım","Hande Dağbaşı","Zafer Yılmaz"]
}

calisan_2 ={
    "calisan_meslek":["Öğretmen","Mimar","Mühendis","Turizm"],
    "calisan_maas":[12000,15500,16750,22560]
}
```

```python
# yukarda oluşturduğumuz sözlükleri ayrı iki d.frame yapalım ardından birleştirelim ...

df_1 = pd.DataFrame(calisan_1)
df_2 = pd.DataFrame(calisan_2)
```

```python
df_1
```

```python
df_2
```

```python
# iki d.frame i birleştirelim ...

pd.concat([df_1,df_2])
```

```python
# axis belirterek birleştirelim

pd.concat(([df_1,df_2]) , axis = 0) # satıra göre 
```

```python
pd.concat(([df_1,df_2]) , axis = 1) # sütuna göre
```
